```python
!pip install great-expectations==0.17.3
from google.colab import drive
drive.mount('/content/drive')

import os
# Set the root directory in Google Drive
root_dir = "/content/drive/My Drive/great_expectations_project/"

# Create the directory if it doesn't exist
if not os.path.exists(root_dir):
    os.makedirs(root_dir, exist_ok=True)
# Change the working directory to the project folder
os.chdir(root_dir)
print("Current working directory:", os.getcwd())

# Pre-req
!great_expectations init
import great_expectations as ge
context = ge.DataContext()

#Load test data
import pandas as pd
df = pd.read_csv('/content/drive/My Drive/data/california_housing_test.csv')
print(df)

#define expectations
from great_expectations.dataset import PandasDataset
dataset = PandasDataset(df)
dataset.expect_column_values_to_be_between('housing_median_age', min_value=0, max_value=100)


#!great_expectations datasource new
!great_expectations suite new

```

```python
print(dir(dataset))
```

```python
# Create an expectation suite
expectation_suite_name = 'my_suite_24'
context.add_expectation_suite(expectation_suite_name)
expectation_suite = context.get_expectation_suite(expectation_suite_name)

# Create a batch
batch = context.create_batch(batch_kwargs=batch_kwargs, data=df)

# Add expectations to the batch
batch.expect_column_to_exist('housing_median_age')
batch.expect_column_values_to_be_unique('median_income')
batch.expect_column_values_to_be_between('housing_median_age', min_value=50, max_value=100)

# Save the expectation suite
context.save_expectation_suite(expectation_suite)

# Create a checkpoint
checkpoint_name = 'my_first_checkpoint'
checkpoint_config = {
    'name': checkpoint_name,
    'expectation_suite_name': expectation_suite_name,
    'batch_kwargs': batch_kwargs
}
context.add_checkpoint(**checkpoint_config)

# Run the checkpoint
results = context.run_checkpoint(checkpoint_name=checkpoint_name)

# Print the results
print(results)
```
